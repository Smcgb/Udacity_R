### Tidiness
#### Define: 
> To merge dataframes using rbind, I will add a City column to each dataframe. For Washington, gender and birth year will be added with default NA values.

> Once data is squared, I will use `plyer rbind.fill` to join all 3 tables into 1 flat file before doing analysis

```r
#creating city columns for ny and chi files
ny$City = 'nyc'
chi$City = 'chi'

#creating columns for wash
wash$Gender = NA
wash$Birth.Year = NA
wash$City = 'wash'
```

```r
df_merge <- rbind(chi, ny)
head(df_merge)
```

```r
dim(df_merge)
```

```r
library("plyr")
df_merge <- rbind.fill(df_merge, wash)
head(df_merge)
```

```r
dim(df_merge)
```

```r
ggplot(aes(x=City), data=df_merge)+
  stat_count(width = 0.5)
```

### Question 3

**Do younger people take longer trips?.**

```r
# Basic scatter plot
ggplot(df_merge, aes(x=Birth.Year, y=Trip.Duration)) + 
  geom_point() +
  stat_smooth()
```

```r
#There is a major outlier here cuasing the data to be skewed.

ggplot(df_merge, aes(x=Birth.Year, y=Trip.Duration)) + 
  geom_boxplot(outlier.colour="red", outlier.shape=8,
                outlier.size=4)
```

```r
boxplot(df_merge$Trip.Duration, plot=FALSE)$out
```

```r
outliers <- boxplot(df_merge$Trip.Duration, plot=FALSE)$out
```

```r
x<-df_merge
x<- x[-which(x$Trip.Duration %in% outliers),]
```

```r
ggplot(x, aes(x=Birth.Year, y=Trip.Duration)) + 
  geom_point() +
  stat_smooth()
```

**Non subs seem to rent bikes for significantly longer trips.**
